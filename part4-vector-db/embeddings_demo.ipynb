```python

import numpy as np
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity


# Sentences
sentences = [
    "The batsman scored a century in the final match",
    "The bowler delivered a perfect yorker to dismiss the player",
    "The team celebrated their victory after a thrilling chase",
    "Fielders positioned themselves strategically to stop boundaries",
    "The chef prepared a delicious pasta with fresh ingredients",
    "Baking a cake requires precise measurements and timing",
    "The soup was simmered slowly to enhance its flavor",
    "Strong passwords help protect accounts from unauthorized access",
    "The firewall blocked suspicious network traffic effectively",
    "Encryption ensures that sensitive data remains secure"
]

# Load model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings
embeddings = model.encode(sentences)

# Cosine similarity matrix
sim_matrix = cosine_similarity(embeddings)

# Plot heatmap
plt.figure(figsize=(8,6))
plt.imshow(sim_matrix, interpolation='nearest')
plt.colorbar()
plt.title("10×10 Cosine Similarity Matrix Heatmap")
plt.xticks(range(10), range(1,11))
plt.yticks(range(10), range(1,11))
plt.show()

# Query sentence
query = "The bowler took three wickets in one over"
query_embedding = model.encode([query])

# Compute similarities
similarities = cosine_similarity(query_embedding, embeddings)[0]

# Top 2 most similar sentences
top_indices = similarities.argsort()[-2:][::-1]

for i in top_indices:
    print(f"Sentence: {sentences[i]}")
    print(f"Similarity Score: {similarities[i]:.4f}\n")
```
